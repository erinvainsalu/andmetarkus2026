```python
def numeric_year(year_string: str) -> int:
    year_number_components = year_string.split(" ")
    year_number_str = year_number_components[0]
    year_era_notation_str = year_number_components[1]
    if year_era_notation_str == "A.D.":
        return int(year_number_str)
    if year_era_notation_str == "B.C.":
        return -int(year_number_str)
    raise ValueError(f"{year_string} does not seem to be a valid year number")
    

print(numeric_year("2024 A.D.")) #2024
print(numeric_year("1 B.C.")) #-1
print(numeric_year("1 BC.")) #-1
```

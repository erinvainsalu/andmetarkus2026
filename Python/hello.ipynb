## if/else

```python
def make_sense_of_this_year(year: int):
    if year < 0:
        print("B.C.")
        if year <= -10000:
            print("Stone age...")
    elif year == 0:
        print("Happy birthday, Jesus!")
    elif year == 1978:
        print("Happy birthday, Mati!")
    else:
        print("A.D.")

make_sense_of_this_year(-90)
```

```python
def make_sense_of_this_year_v2(year: int) -> str:
    if year < 0:
        tekst = "B.C."
        if year <= -10000:
            tekst = "Stone age..."
    elif year == 0:
        tekst = "Happy birthday, Jesus!"
    elif year == 1978:
        tekst = "Happy birthday, Mati!"
    else:
        tekst = "A.D."
    return tekst

make_sense_of_this_year_v2(-10000)
```

```python
def print_year(year: int):
    if year < 0:
        print(f"{abs(year)} B.C.")
    else:
        print(f"{year} A.D.")

print_year(-90)
print_year(2025)
```

```python
def print_year(year: int) -> str:
    if year < 0:
        return f"{abs(year)} B.C."
    else:
        return f"{year} A.D."

print(print_year(-90))
print(print_year(2025))
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['target'] = iris_data.target
iris['species'] = iris['target'].map(dict(enumerate(iris_data.target_names)))
iris.pop('target')
# rename funktsioon tagastab uue tabeli
iris = iris.rename(columns={
    "sepal length (cm)": "sepal_length",
    "sepal width (cm)": "sepal_width",
    "petal length (cm)": "petal_length",
    "petal width (cm)": "petal"
})
iris
```

```python
duckdb.sql("""
    SELECT
        species, 
        AVG(petal_length) AS avg_petal_length
    FROM iris
    GROUP BY species
    ORDER BY avg_petal_length DESC
""").df()
```

```python
def year_display(year: int) -> str:
    if year < 0:
        return f"{abs(year)} B.C."
    elif year > 0:
        return f"{year} A.D."
    else:
        return "Palju õnne, Jeesus!"

print(year_display(-10000))
print(year_display(2024))
print(year_display(0))
```
